```python
def process_mmc_table_init(spark, db, table_names):
    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{vin_meid_mapping} (
        vin  STRING,
        ciamid STRING
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        vin_meid_mapping=table_names["vin_meid_mapping"]
    ))
        
    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_product} (
        product_id  STRING,
        variant_id STRING,
        product_name  STRING,
        product_type  STRING,
        msrp DOUBLE,
        purchase_price DOUBLE
    )
    USING DELTA
    """.format(
        db = db,
        mmc_product=table_names["mmc_product"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_coupons} (
        vin STRING,
        promotion_code STRING,
        coupon_code STRING,
        start_date DATE,
        end_date DATE,
        redeemed_date DATE,
        status STRING
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_coupons=table_names["mmc_coupons"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_product_attributes} (
        product_id STRING,
        product_type STRING,
        card_uv_imps INT, 
        pdp_uv_imps INT, 
        tot_order_cnt INT,
        app_order_cnt INT,
        avg_order_msrp DOUBLE,
        avg_order_baseprice DOUBLE,
        avg_trans_price DOUBLE
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_product_attributes=table_names["mmc_product_attributes"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_user_attributes} (
        user_id STRING,
        store_visits INT,
        platform STRING, 
        pdp_views INT, 
        tot_items_ordered INT,
        app_items_ordered INT,
        coupons_owned INT,
        coupons_redeemed INT
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_user_attributes=table_names["mmc_user_attributes"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_user_product_interaction} (
        user_id STRING, 
        product_id STRING, 
        card_impressions INT,
        details_viewed INT,
        orders_initiated INT,
        orders_placed INT,
        total_payment DOUBLE
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_user_product_interaction=table_names["mmc_user_product_interaction"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_user_features} (
        user_id STRING, 
        platform STRING, 
        store_visits_30d INT,
        store_visits_90d INT,
        pdp_views_30d INT,
        pdp_views_90d INT,
        tot_items_ordered_30d INT,
        tot_items_ordered_90d INT,
        tot_items_ordered_180d INT,
        app_items_ordered_30d INT,
        app_items_ordered_90d INT,
        app_items_ordered_180d INT,
        coupons_owned_30d INT,
        coupons_owned_90d INT,
        coupons_owned_180d INT,
        coupons_redeemed_30d INT,
        coupons_redeemed_90d INT,
        coupons_redeemed_180d INT
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_user_features=table_names["mmc_user_features"]
    ))

    spark.sql("""
    CREATE TABLE IF NOT EXISTS {db}.{mmc_product_features} (
        product_id STRING, 
        product_type STRING,
        card_uv_imps_30d INT,
        card_uv_imps_90d INT,
        pdp_uv_imps_30d INT,
        pdp_uv_imps_90d INT,
        tot_order_cnt_30d INT,
        tot_order_cnt_90d INT,
        app_order_cnt_30d INT,
        app_order_cnt_90d INT,
        avg_order_msrp DOUBLE,
        avg_order_baseprice_60d DOUBLE,
        avg_trans_price_30d DOUBLE,
        avg_trans_price_90d DOUBLE
    )
    USING DELTA
    PARTITIONED BY (
        partition_date  STRING
    )
    """.format(
        db = db,
        mmc_product_features=table_names["mmc_product_features"]
    ))
```

```python
import json
config_path = dbutils.widgets.get("table_config_path")
with open(config_path, "r") as file:
    config = json.load(file)
db, table_names = config['DATABASE'], config['TABLE_NAMES']
```

```python
process_mmc_table_init(spark, db, table_names)
```
